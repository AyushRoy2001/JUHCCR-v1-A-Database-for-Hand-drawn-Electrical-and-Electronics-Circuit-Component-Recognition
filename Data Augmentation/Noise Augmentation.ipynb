```python
img_dir = "/content/drive/MyDrive/Data/20/*"
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def addNoise(img,c):
    dst = np.empty_like(img)
    noise = cv2.randn(dst, (0,0,0), (20,20,20))
    img_noise = cv2.addWeighted(img, 0.5, noise, 0.5, 30)

    cv2_imshow(img_noise)
    cv2.imwrite('/content/drive/MyDrive/Data/20/train/componentlnoise%d.bmp' % c, img_noise)
    
```

```python
# import the library opencv
import cv2
# globbing utility.
import glob
c=0
for file in glob.glob(img_dir):
  image_read = cv2.imread(file)
  c=c+1
  addNoise(image_read,c)
```
